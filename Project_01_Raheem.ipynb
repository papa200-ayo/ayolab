```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_table(filepath_or_buffer = 'auto_data.txt', sep='\t')
weight = list(df.wt)
mpg = list(df.mpg)
ln_mpg = list(df.ln_mpg)
print('Length of list weight : ' + str(len(weight)))
print('Length of list mpg : ' + str(len(mpg)))
print('Length of list ln_mpg : ' + str(len(ln_mpg)))
print("{:>6}{:>8}{:>10}" .format("weight" ,"mpg","ln_mpg"))
print("{:-^24}" .format(""))
for i in  range(len(weight)):
    print("{:>6}{:>8}{:>10}".format (weight[i],mpg[i],ln_mpg[i]))
```

# Part 3: Descriptive Statistics

```python
#List for the values of x and y
x_train=[2.5,2.50,2.75,3.0,3.25]
y_train=[2.6,2.8,3.0,3,2,3.4,3.6]
n=len(x_train)   #Finding the length of list
X=sum(x_train)   # Calculating the sum of list x
Y=sum(y_train)   # Calculating the sum of list y
Mean_x=X/n       # Calculating the mean of x  
Mean_y=Y/n        # Calculating the mean of y
format_float1 = "{:.2f}".format(Mean_x)   # 2 decimal places
format_float2 = "{:.4f}".format(Mean_y)   #4 decimal places
print("Mean of x: ",format_float1)
print("Mean of y: ",format_float2)
()
Sxx=0.0                      
for i in range(len(x_train)): # loop for squared deviation x
  Sxx=Sxx+((x_train[i]-Mean_x)**2)
Syy=0.0                       #Initialising sum of Syy as 0
for i in range(len(y_train)):# loop for squared deviation y
  Syy=Syy+((y_train[i]-Mean_y)**2)
format_float3 = "{:.2f}".format(Sxx)
format_float4 = "{:.4f}".format(Syy)
print("Sum of the squared deviations for X: ",format_float3)
print("Sum of the squared deviations for Y: ",format_float4)
var_x=Sxx/(n-1) # variance x
var_y=Syy/(n-1)# variance y
format_float5 = "{:.2f}".format(var_x)
format_float6 = "{:.4f}".format(var_y)
print("Variance of X: ",format_float5)
print("Variance of Y: ",format_float6)
```
